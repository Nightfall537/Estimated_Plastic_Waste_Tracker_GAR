```python
#Interactive Plastic Waste Dashboard of Unrecycled Plastic Waste and Plastic Waste Generated from 2025 to 2035

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
import pandas as pd  # Import pandas

# Load the data
data = {
    'Years': [2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032, 2033, 2034, 2035],
    'Plastic Waste Generated (tonnes)': [109500, 126165, 129509, 132865, 136231, 139600, 142984, 146377, 149781, 153193, 156608],
    'Unrecycled Plastic Waste (tonnes)': [99166, 111151, 111119, 110809, 111035, 109865, 109097, 108173, 107243, 106010, 104614]
}
df = pd.DataFrame(data)

# Calculate Recycled Plastic Waste
df['Recycled Plastic Waste (tonnes)'] = df['Plastic Waste Generated (tonnes)'] - df['Unrecycled Plastic Waste (tonnes)']

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div(children=[
    html.H1(
        children='Plastic Waste Management in Ghana',
        style={
            'textAlign': 'center',
            'color': '#2C3E50'  # Darker, more professional color
        }
    ),

    html.Div(
        children='This dashboard shows the amount of plastic waste generated, recycled, and unrecycled in Ghana from 2025 to 2035.',
        style={
            'textAlign': 'center',
            'color': '#7F8C8D',  # Muted color for description
            'marginBottom': '20px'
        }
    ),

    dcc.Graph(
        id='plastic-waste-bar-chart',
        style={
            'width': '90%',  # Responsive width
            'margin': 'auto'  # Center the graph
        },
        figure={}  # Important:  Initialize the figure as an empty dict!
    ),
])

# Define callback to update the chart
@app.callback(
    Output('plastic-waste-bar-chart', 'figure'),
    [Input('plastic-waste-bar-chart', 'relayoutData')]  # Add relayoutData as input
)
def update_chart(relayout_data):
    # Create the stacked bar chart
    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=df['Years'],
        y=df['Plastic Waste Generated (tonnes)'],
        name='Plastic Waste Generated (tonnes)',
        marker_color='#8884d8'
    ))
    fig.add_trace(go.Bar(
        x=df['Years'],
        y=df['Recycled Plastic Waste (tonnes)'],
        name='Recycled Plastic Waste (tonnes)',
        marker_color='#82ca9d'
    ))
    fig.add_trace(go.Bar(
        x=df['Years'],
        y=df['Unrecycled Plastic Waste (tonnes)'],
        name='Unrecycled Plastic Waste (tonnes)',
        marker_color='#ffc658'
    ))

    # Update the layout
    fig.update_layout(
        title='Plastic Waste Overview',
        xaxis_title='Years',
        yaxis_title='Amount of Plastic Waste (tonnes)',
        barmode='stack',
        # Add hover labels
        hoverlabel=dict(
            bgcolor="white",
            font_size=13,
            font_family="Arial"
        ),
        # Make it responsive
        margin=dict(l=20, r=20, t=40, b=20),
        # Add interactivity
        dragmode="zoom",  # Enable zoom and pan
        hovermode="x unified",  # improve hover
        # Add a legend title
        legend=dict(
            title="Waste Type",
            traceorder="normal",
            font=dict(
                family="sans-serif",
                size=12,
                color="black"
            ),
            bgcolor="rgba(0,0,0,0)",  # make legend background transparent
            bordercolor="rgba(0,0,0,0)",
        )
    )
    return fig

if __name__ == '__main__':
    app.run(debug=True, port=8050) # Changed app.run_server to app.run
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
import pandas as pd

# Load the data
data = {
    'Years': [2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032, 2033, 2034, 2035],
    'Population in GAR': [5455692, 6286269, 6452843, 6620120, 6787804, 6955697, 7124253, 7293371, 7462978, 7632950, 7803070],
    'Plastic Waste Generated (tonnes)': [109500, 126165, 129509, 132865, 136231, 139600, 142984, 146377, 149781, 153193, 156608],
}
df = pd.DataFrame(data)


# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div(children=[
    html.H1(
        children='Plastic Waste Management and Population in GAR',
        style={
            'textAlign': 'center',
            'color': '#2C3E50'  # Darker, more professional color
        }
    ),

    html.Div(
        children='This dashboard shows the estimated population in the Greater Accra Region and the amount of plastic waste generated from 2025 to 2035.',
        style={
            'textAlign': 'center',
            'color': '#7F8C8D',  # Muted color for description
            'marginBottom': '20px'
        }
    ),

    dcc.Graph(
        id='population-and-waste-chart',
        style={
            'width': '90%',
            'margin': 'auto'
        },
        figure={}
    ),
])


# Define callback to update the chart
@app.callback(
    Output('population-and-waste-chart', 'figure'),
    [Input('population-and-waste-chart', 'relayoutData')]
)
def update_chart(relayout_data):
    # Create the figure with two y-axes
    fig = go.Figure()

    # Add trace for Population in GAR
    fig.add_trace(go.Scatter(
        x=df['Years'],
        y=df['Population in GAR'],
        name='Population in GAR',
        marker_color='#1E88E5',  # Blue color
        yaxis='y1'
    ))

    # Add trace for Plastic Waste Generated (tonnes)
    fig.add_trace(go.Bar(
        x=df['Years'],
        y=df['Plastic Waste Generated (tonnes)'],
        name='Plastic Waste Generated (tonnes)',
        marker_color='#8884d8',  # Purple
        yaxis='y2'
    ))

    # Update layout to include titles and multiple y-axes
    fig.update_layout(
        title='Population in GAR and Plastic Waste Generated (2025-2035)',
        xaxis_title='Year',
        yaxis_title='Population in GAR',  # Title for the first y-axis
        yaxis2=dict(
            title='Plastic Waste Generated (tonnes)',  # Title for the second y-axis
            overlaying='y',
            side='right'
        ),
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="right",
            x=1
        ),
        # Make it responsive
        margin=dict(l=20, r=20, t=40, b=20),
        dragmode="zoom",  # Enable zoom and pan
        hovermode="x unified"
    )

    return fig


if __name__ == '__main__':
    app.run(debug=True, port=8052)
```
